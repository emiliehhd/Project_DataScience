## Nettoyage du dataset

Suppression des colonnes inutiles. Les colonnes qui seront supprimées continnent des informations qui ne seront pas utilisées pour la modélisation, tel que:
 - les liens et images (`player_url`, `player_face_url`, ..)
 - les IDs (`sofifa_id`)
 - les informations spécifiques / administratives(`real_face`, `body_type`, ..)

```python
cols_a_supprimer = [
    'player_url', 'player_face_url',
    'club_logo_url', 'club_flag_url', 'nation_logo_url', 'nation_flag_url',
    'sofifa_id',
    'real_face', 'body_type',
    'long_name', 'dob', 'club_team_id',
    'club_jersey_number', 'nationality_id', 'nation_team_id',
    'nation_jersey_number', 'nation_position', 'player_tags',
    'player_traits',
    'ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw',
    'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm',
    'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb',
    'cb', 'rcb', 'rb', 'gk'
]

df = df.drop(columns=cols_a_supprimer)

print(f"Colonnes restantes : {df.shape[1]}")
print(f"Joueurs : {df.shape[0]}")
print("\nListe finale des colonnes :")
print(list(df.columns))
```

## Valeurs manquantes

```python
# Analyse des valeurs manquantes
valeurs_manquantes = df.isnull().sum()
valeurs_manquantes_pct = (df.isnull().sum() / len(df) * 100).round(2)

# tableau récapitulatif (décroissant)
missing_df = pd.DataFrame({
    'Valeurs manquantes': valeurs_manquantes,
    'Pourcentage (%)': valeurs_manquantes_pct
}).sort_values('Pourcentage (%)', ascending=False)

# colonnes qui ont au moins 1 valeur manquante
missing_df = missing_df[missing_df['Valeurs manquantes'] > 0]

print(f"Colonnes avec valeurs manquantes : {len(missing_df)}")
print()
print(missing_df)
```

À supprimer — trop de valeurs manquantes
club_loaned_from → 94.27% manquant. Normal : seuls les joueurs en prêt ont cette info. Inutilisable, on supprime.
goalkeeping_speed → 88.92% manquant. Cette stat n'existe que pour les gardiens. On supprime.

Cas particulier — les 11.08% (pace, shooting, passing, dribbling, defending, physic)
Ces 6 colonnes ont exactement le même nombre de valeurs manquantes — 2132 joueurs. Ce n'est pas un hasard : ce sont les gardiens de but ! FIFA ne leur attribue pas ces stats générales. On a deux options :

Les supprimer → on perd de l'info sur les joueurs de champ
Les remplir par 0 pour les gardiens → logique métier : un gardien n'a pas de stat "shooting"

👉 On choisit de remplir par 0, c'est justifiable et on garde l'info.

🔧 À traiter intelligemment — les petits pourcentages
release_clause_eur (6.11%) et club_joined (6.05%) → on remplace par la médiane pour la clause et par une valeur neutre pour la date.
value_eur (0.38%) → c'est notre variable cible ! On ne peut pas inventer une valeur. On supprime ces 74 lignes.
Les 61 joueurs manquants sur club_name, league_name, league_level, club_position, club_contract_valid_until, wage_eur → ce sont probablement des joueurs sans club. On les supprime car ils représentent moins de 0.32% des données.
