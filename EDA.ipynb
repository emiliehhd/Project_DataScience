À supprimer — trop de valeurs manquantes
club_loaned_from → 94.27% manquant. Normal : seuls les joueurs en prêt ont cette info. Inutilisable, on supprime.
goalkeeping_speed → 88.92% manquant. Cette stat n'existe que pour les gardiens. On supprime.

Cas particulier — les 11.08% (pace, shooting, passing, dribbling, defending, physic)
Ces 6 colonnes ont exactement le même nombre de valeurs manquantes — 2132 joueurs. Ce n'est pas un hasard : ce sont les gardiens de but ! FIFA ne leur attribue pas ces stats générales. On a deux options :

Les supprimer → on perd de l'info sur les joueurs de champ
Les remplir par 0 pour les gardiens → logique métier : un gardien n'a pas de stat "shooting"

👉 On choisit de remplir par 0, c'est justifiable et on garde l'info.

🔧 À traiter intelligemment — les petits pourcentages
release_clause_eur (6.11%) et club_joined (6.05%) → on remplace par la médiane pour la clause et par une valeur neutre pour la date.
value_eur (0.38%) → c'est notre variable cible ! On ne peut pas inventer une valeur. On supprime ces 74 lignes.
Les 61 joueurs manquants sur club_name, league_name, league_level, club_position, club_contract_valid_until, wage_eur → ce sont probablement des joueurs sans club. On les supprime car ils représentent moins de 0.32% des données.

```python
# Supprimer les colonnes trop incomplètes 
df = df.drop(columns=['club_loaned_from', 'goalkeeping_speed'])

# ── 2. Remplir les stats générales des gardiens par 0 ───────────────────
cols_stats_generales = ['pace', 'shooting', 'passing', 
                        'dribbling', 'defending', 'physic']
df[cols_stats_generales] = df[cols_stats_generales].fillna(0)

# ── 3. Remplir release_clause_eur par la médiane ────────────────────────
mediane_clause = df['release_clause_eur'].median()
df['release_clause_eur'] = df['release_clause_eur'].fillna(mediane_clause)

# ── 4. Supprimer les lignes sans valeur cible (value_eur) ───────────────
df = df.dropna(subset=['value_eur'])

# ── 5. Supprimer les joueurs sans club (61 lignes) ──────────────────────
df = df.dropna(subset=['club_name'])

# ── 6. Remplir club_joined par la valeur la plus fréquente ──────────────
df['club_joined'] = df['club_joined'].fillna(df['club_joined'].mode()[0])

# ── Vérification finale ──────────────────────────────────────────────────
print(f"Dataset final : {df.shape[0]} joueurs | {df.shape[1]} colonnes")
print(f"Valeurs manquantes restantes : {df.isnull().sum().sum()}")
```

```python
# Calcul de la skewness sur toutes les colonnes numériques
skewness = df.select_dtypes(include='number').skew().sort_values(ascending=False)

# Tableau récapitulatif avec interprétation
skew_df = pd.DataFrame({
    'Skewness': skewness.round(2)
})

def interpreter_skew(val):
    if abs(val) <= 0.5:
        return '✅ Symétrique'
    elif abs(val) <= 1:
        return '⚠️ Légèrement asymétrique'
    else:
        return '🚨 Fortement asymétrique'

skew_df['Interprétation'] = skew_df['Skewness'].apply(interpreter_skew)

print("=== Skewness des variables numériques ===\n")
print(skew_df.to_string())

# Compter les cas problématiques
n_problematiques = (skew_df['Skewness'].abs() > 1).sum()
print(f"\n Variables fortement asymétriques (|skew| > 1) : {n_problematiques}")
```

```python
# Transformation logarithmique sur les variables financières
cols_log = ['value_eur', 'wage_eur', 'release_clause_eur']

for col in cols_log:
    df[f'{col}_log'] = np.log1p(df[col])
    print(f"✓ {col}_log créée | skewness avant : {df[col].skew():.2f} "
          f"→ après : {df[f'{col}_log'].skew():.2f}")

# Visualisation avant / après
fig, axes = plt.subplots(3, 2, figsize=(14, 12))

for i, col in enumerate(cols_log):
    # Avant transformation
    axes[i, 0].hist(df[col], bins=50, color='coral', edgecolor='white')
    axes[i, 0].set_title(f'{col} — AVANT (skew={df[col].skew():.2f})')
    axes[i, 0].set_xlabel('Valeur')

    # Après transformation
    axes[i, 1].hist(df[f'{col}_log'], bins=50, 
                    color='steelblue', edgecolor='white')
    axes[i, 1].set_title(f'{col}_log — APRÈS (skew={df[f"{col}_log"].skew():.2f})')
    axes[i, 1].set_xlabel('Log(Valeur)')

plt.suptitle('Effet de la transformation logarithmique', 
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
print("=== Outliers AVANT transformation log ===\n")
for col in ['value_eur', 'wage_eur', 'release_clause_eur']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    borne_sup = Q3 + 1.5 * IQR
    outliers = df[df[col] > borne_sup]
    print(f"{col:<25} → {len(outliers):>4} outliers ({len(outliers)/len(df)*100:.1f}%)")

print("\n=== Outliers APRÈS transformation log ===\n")
for col in ['value_eur_log', 'wage_eur_log', 'release_clause_eur_log']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    borne_sup = Q3 + 1.5 * IQR
    outliers = df[df[col] > borne_sup]
    print(f"{col:<29} → {len(outliers):>4} outliers ({len(outliers)/len(df)*100:.1f}%)")
```

## Séparation des features (X) et de la variable cible (y)

```python
# Notre cible : value_eur_log (version transformée)
y = df['value_eur_log']

# On supprime les colonnes qu'on ne veut pas dans les features
cols_a_exclure = [
    'value_eur',           # version brute de la cible
    'value_eur_log',       # notre cible
    'wage_eur',            # version brute (on garde wage_eur_log)
    'release_clause_eur',  # version brute (on garde release_clause_eur_log)
    'player_positions',    # remplacée par poste_principal
    'short_name',          # identifiant textuel inutile
]

X = df.drop(columns=cols_a_exclure)

print(f"✓ Features (X) : {X.shape[1]} colonnes | {X.shape[0]} joueurs")
print(f"✓ Cible (y)    : value_eur_log")
print(f"\nColonnes de X :")
print(list(X.columns))
```

## Encoding des vars catégorielles

D'abord, identifier les catégorielles
On a 5 variables textuelles dans X :
- ColonneNb valeurs uniquesTraitementpreferred_foot2 (Left/Right): Label encoding
- work_rate~9 valeurs : Label encoding
- poste_principal~20 postes : One-hot encoding
- club_name~700 clubs : Target encoding
- league_name~30 ligues  : One-hot encoding
- nationality_name~160 nationalités : Target encoding
- club_position~30 positions : One-hot encoding

💡 Pourquoi ces choix ?

- Label encoding → pour les variables avec peu de valeurs ordonnées ou binaires
- One-hot encoding → pour les catégories sans ordre et peu nombreuses
- Target encoding → pour les catégories très nombreuses (club_name, nationality_name) — on remplace chaque catégorie par la moyenne de la cible correspondante, ce qui évite d'exploser le nombre de colonnes


```python
from sklearn.preprocessing import LabelEncoder

# ── 1. Label Encoding — preferred_foot ──────────────────────────────────
le = LabelEncoder()
X['preferred_foot'] = le.fit_transform(X['preferred_foot'])
print("✓ preferred_foot encodé (Right=1, Left=0)")

# ── 2. Label Encoding — work_rate ───────────────────────────────────────
X['work_rate'] = le.fit_transform(X['work_rate'])
print("✓ work_rate encodé")

# ── 3. Target Encoding — club_name ──────────────────────────────────────
club_target_mean = df.groupby('club_name')['value_eur_log'].mean()
X['club_name'] = X['club_name'].map(club_target_mean)
print("✓ club_name encodé par target encoding")

# ── 4. Target Encoding — nationality_name ───────────────────────────────
nation_target_mean = df.groupby('nationality_name')['value_eur_log'].mean()
X['nationality_name'] = X['nationality_name'].map(nation_target_mean)
print("✓ nationality_name encodé par target encoding")

# ── 5. One-Hot Encoding — poste_principal, league_name, club_position ───
X = pd.get_dummies(X, columns=['poste_principal', 
                                'league_name', 
                                'club_position'], 
                   drop_first=True)
print("✓ poste_principal, league_name, club_position encodés en one-hot")

# ── Vérification ─────────────────────────────────────────────────────────
print(f"\n✓ Shape finale de X : {X.shape}")
print(f"✓ Valeurs manquantes dans X : {X.isnull().sum().sum()}")
print(f"✓ Types restants :")
print(X.dtypes.value_counts())
```
